# Jaguar

Display the **mean** of Jaguar

```python
j_mean = df["Jaguar"].mean()
print("Jaguar mean = "+str(j_mean))
```

Display the **median** of Jaguar 

```python
j_median = df['Jaguar'].median()
print("Jaguar median = " + str(j_median))
```

Display the **mode** of Jaguar 

```python
j_mode = df["Jaguar"].mode().to_numpy()[0]
print("Jaguar mode = " + str(j_mode))
```

Return **Range**

```python
j_range = df["Jaguar"].max() - df["Jaguar"].min()
print("Jaguar range = " +str(j_range))
```

Return the **Variance**

```python
j_var = df['Jaguar'].var()
print('Jaguar variance = ' + str(j_var.round(2)))
```

Return the **Standard Deviation**

```python
j_sd = df['Jaguar'].std()
print('Jaguar standard deviation = ' + str(j_sd.round(2)))
```

Return the **Coefficient of Variation**

```python
j_cv = j_sd/df['Jaguar'].mean()
print("Jaguar coefficient of variation = " + str(j_cv.round(2)))
```

Visualize the Measure of Variability using **Boxplot**

```python
plt.boxplot(df['Jaguar'])
plt.title("Jaguar Performance Boxplot")
plt.xlabel("Jaguar")
plt.ylabel("Value")
plt.xticks([])
plt.show()
```

# Data Analysis Between Jaguar and Panther

Display the data using boxplot

**Mean**

Jaguar has higher mean which indicates larger data value compared to Panther.

```python
x = ["Jaguar", "Panther"]
y = [1087.80, 1005.33]

plt.bar(x, y, color='pink')


for i, value in enumerate(y):
    plt.text(i, value + 10, f"{value:.2f}", ha='center', fontsize=10)

plt.title('Mean Comparison of Jaguar and Panther Performance')
plt.xlabel('Mean')
plt.ylabel('Value')

plt.show()
```

**Median**

In this comparison, Jaguar still has higher median compared to Panther.

```python
x = ["Jaguar", "Panther"]
y = [998, 994.50]

plt.bar(x, y, color='pink')


for i, value in enumerate(y):
    plt.text(i, value + 10, f"{value:.2f}", ha='center', fontsize=10)

plt.title('Median Comparison of Jaguar and Panther Performance')
plt.xlabel('Mean')
plt.ylabel('Value')

plt.show()
```

**Variance**

```python

jaguar_data = np.random.normal(0, 80749.82, 1000)
panther_data = np.random.normal(0, 27642.85, 1000)


plt.figure(figsize=(10,6))
plt.hist(jaguar_data, bins=30, alpha=0.5, label='Jaguar', color='pink')
plt.hist(panther_data, bins=30, alpha=0.5, label='Panther', color='green')


plt.xlabel('Data Points')
plt.ylabel('Frequency')
plt.title('Histogram of Jaguar and Panther Variance')
plt.legend()

# Display the plot
plt.show()

#reference: https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html
#reference: https://www.geeksforgeeks.org/matplotlib-pyplot-hist-in-python/
```

**Standard Deviation**

Jaguar showed significant standard deviation with **284.17** compared to Panther with **166.26**

```python
jaguar_data = np.random.normal(0, 284.17, 1000)
panther_data = np.random.normal(0, 166.26, 1000)


plt.figure(figsize=(10,6))
plt.hist(jaguar_data, bins=30, alpha=0.5, label='Jaguar', color='pink')
plt.hist(panther_data, bins=30, alpha=0.5, label='Panther', color='green')


plt.xlabel('Data Points')
plt.ylabel('Frequency')
plt.title('Histogram of Jaguar and Panther Standard Deviations')
plt.legend()

# Display the plot
plt.show()

#reference: https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html
#reference: https://www.geeksforgeeks.org/matplotlib-pyplot-hist-in-python/
```

**Coefficient of Variation**

The coefficient of variation (CV) further highlights this difference; Jaguar (0.26) has more relative variability than Panther (0.17).

```python
jaguar_data = np.random.normal(0, 0.26, 1000)
panther_data = np.random.normal(0, 0.17, 1000)


plt.figure(figsize=(10,6))
plt.hist(jaguar_data, bins=30, alpha=0.5, label='Jaguar', color='pink')
plt.hist(panther_data, bins=30, alpha=0.5, label='Panther', color='green')


plt.xlabel('Data Points')
plt.ylabel('Frequency')
plt.title('Histogram of Jaguar and Panther CV')
plt.legend()

# Display the plot
plt.show()

#reference: https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html
#reference: https://www.geeksforgeeks.org/matplotlib-pyplot-hist-in-python/
```
